```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
import seaborn as sb
```

```python
sb.scatterplot(data = df, y = 'population', x = 'households')
```

```python
sb.relplot(data = df, y = 'median_house_value', x = 'longitude', kind = 'line')
```

```python
sb.histplot(data = df, x = 'housing_median_age', bins = 100)
```

```python
sb.histplot(data = df, x = 'median_house_value', bins = 50)
```

```python
df.loc[df['housing_median_age'] >= 30,'age'] = 'old'
df.loc[(df['housing_median_age'] < 30) & (df['housing_median_age'] >= 10) ,'age'] = 'medium'
df.loc[df['housing_median_age'] < 10,'age'] = 'young'
```

```python
df.head()
```

```python
sb.histplot(data = df, x = 'median_house_value', bins = 50, hue = 'age')
```

```python
df.loc[df['median_house_value'] > 490000, 'price'] = 'expensive'
df.loc[df['median_house_value'] <= 490000, 'price'] = 'cheap'
```

```python
df.head()
```

```python
sb.scatterplot(data = df, x = 'latitude', y = 'longitude', hue = 'price')
```

№10.3[67]. Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.

high - длинный(от 42)

middle - средний(от 35 до 42)

low - маленький(до 35)

```python
pg.loc[pg['bill_length_mm'] >= 42, 'length'] = 'high'
pg.loc[(pg['bill_length_mm'] >= 35) & (pg['bill_length_mm'] < 42), 'length'] = 'middle'
pg.loc[pg['bill_length_mm'] <35, 'length'] = 'low'
```
